```python
# Read the CSV file from Google Drive
data_weather = pd.read_csv("/Users/ahmedalqubaisi/Desktop/KU/Classes/709 Stochastic process emilio/LuxWeather.csv")

# Display the first few rows of the DataFrame
print(data_weather.head())
```

```python
# prompt: drop the first three rows of data_weather dataframe and reset index

data_weather.drop([0, 1, 2], axis=0, inplace=True)
data_weather.reset_index(drop=True, inplace=True)
```

```python
# prompt: rename the first column to 'timestamp' and second column to 'temp'. then remove the rest of the columns. then remove the rest of the NaN rows

data_weather.rename(columns={data_weather.columns[0]: "timestamp", data_weather.columns[1]: "temp"}, inplace=True)
data_weather.drop(data_weather.columns[2:], axis=1, inplace=True)
data_weather.dropna(axis=0, how='any', inplace=True)
print(data_weather.head())
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data_weather['timestamp'] = pd.to_datetime(data_weather['timestamp'])
```

```python

# Merge the two DataFrames based on the datetime column
merged_data = pd.merge(data, data_weather, left_on='timestamp', right_on='timestamp', how='left')


# Print the merged DataFrame
print(merged_data)
```

```python
merged_data['temp'] = pd.to_numeric(merged_data['temp'])
def g(x):
    return 0.01*x**2 - 0.17*x + 1.57

tau_values = np.log(1 + g(merged_data['temp']))

# Adjust the value of k
k = 1 - tau_values.mean()

# Add the tau column to the merged DataFrame
merged_data['tau'] = tau_values + k

# Print the updated DataFrame
print(merged_data)

# Check the range of the tau column
tau_min = merged_data['tau'].min()
tau_max = merged_data['tau'].max()

# Print the minimum and maximum values of tau
print("Minimum Tau:", tau_min)
print("Maximum Tau:", tau_max)
print(k)
```

```python
# prompt: graph tau over time

plt.figure(figsize=(12, 6))
plt.plot(merged_data['timestamp'], merged_data['tau'], label='Tau')
plt.xlabel('Timestamp')
plt.ylabel('Tau Value')
plt.title('Tau Values over Time')
plt.grid(True)
plt.show()
```

```python
# prompt: graph the temperature over time

plt.figure(figsize=(12, 6))
plt.plot(merged_data['timestamp'], merged_data['temp'], label='Temperature')
plt.xlabel('Timestamp')
plt.ylabel('Temperature')
plt.title('Temperature over Time')
plt.grid(True)
plt.show()
```

```python
# prompt: add a column to merged data that labels the date as either weekday or weekend. where weekday is given a 1 and weekend given a 0

def label_weekday(row):
  if row['day_of_week'] in ['Saturday', 'Sunday']:
    return 0
  else:
    return 1

merged_data['weekday_weekend'] = merged_data.apply(label_weekday, axis=1)
```

```python
# Assuming your datetime column is named 'timestamp'
merged_data['timestamp'] = pd.to_datetime(merged_data['timestamp'])

# Generate 'hour', 'weekday', and 'month' features
merged_data['hour'] = merged_data['timestamp'].dt.hour
merged_data['weekday'] = merged_data['timestamp'].dt.weekday  # Monday=0, Sunday=6
merged_data['month'] = merged_data['timestamp'].dt.month

# Print the updated DataFrame to verify
print(merged_data.head())
```

```python


# Filter the dataframe to include only dates up to March 2021
merged_data_filtered = merged_data[merged_data['timestamp'] <= '2021-03-31']

# Check the first few rows of the filtered dataframe
print(merged_data_filtered.head())
```

```python

merged_data_filtered['timestamp'] = pd.to_datetime(merged_data_filtered['timestamp'])

# Plot electricity_price and temp over time
plt.figure(figsize=(10, 6))
plt.plot(merged_data_filtered['timestamp'], merged_data_filtered['electricity_price'], label='Electricity Price')
plt.plot(merged_data_filtered['timestamp'], merged_data_filtered['temp'], label='Temperature')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Electricity Price and Temperature Over Time')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# prompt: create an X and y from the merged data, where X are the feature columns and y is just the electricity price

# Define features and target
features = ['hour', 'weekday', 'month', 'temp', 'tau', 'weekday_weekend']
target = 'electricity_price'

# Create X and y
X = merged_data_filtered[features]
y = merged_data_filtered[target]
```

```python
import numpy as np
import tensorflow as tf

# Define the split ratios
train_ratio = 0.7
val_ratio = 0.1
test_ratio = 0.2

# Calculate the split indices
total_samples = len(X)
train_end_idx = int(total_samples * train_ratio)
val_end_idx = int(total_samples * (train_ratio + val_ratio))

# Split the data into training, validation, and test sets
X_train, y_train = X[:train_end_idx], y[:train_end_idx]
X_val, y_val = X[train_end_idx:val_end_idx], y[train_end_idx:val_end_idx]
X_test, y_test = X[val_end_idx:], y[val_end_idx:]


```

```python
from sklearn.preprocessing import MinMaxScaler

# Create MinMaxScaler instances for X and y
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

# Fit the scalers on the training data
scaler_X.fit(X_train)
scaler_y.fit(y_train.values.reshape(-1, 1))  # Reshape y_train for compatibility

# Transform the training, validation, and test data
X_train_scaled = scaler_X.transform(X_train)
y_train_scaled = scaler_y.transform(y_train.values.reshape(-1, 1))

X_val_scaled = scaler_X.transform(X_val)
y_val_scaled = scaler_y.transform(y_val.values.reshape(-1, 1))

X_test_scaled = scaler_X.transform(X_test)
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))

```

```python
import tensorflow as tf
tf.keras.backend.clear_session()
# Define the feedforward neural network model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer (1 neuron for regression)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse','mae'])

# Train the model
history = model.fit(X_train_scaled, y_train_scaled, epochs=50, batch_size=16, validation_data=(X_val_scaled, y_val_scaled))

# Evaluate the model on the test data
loss = model.evaluate(X_test_scaled, y_test_scaled)

# Print the evaluation metrics
print("Test Loss:", loss)

```

```python

# Plot training loss
plt.plot(history.history['loss'], label='Training Loss')
# Plot validation loss
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training MAE
plt.plot(history.history['mae'], label='Training MAE')
# Plot validation MAE
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()
```

```python

# Make predictions on the test data
forecast = model.predict(X_test_scaled)

# Inverse transform the scaled predictions to get the original values
forecast_inverse = scaler_y.inverse_transform(forecast)

# Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, forecast_inverse))
print('RMSE of forecast:', round(rmse, 3))
```

```python
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Dense, Flatten

# Clear the previous session to release resources
tf.keras.backend.clear_session()

# Define the number of time steps
time_steps = 24

# Determine the number of samples and remaining data after discarding incomplete sequences
num_samples = len(X_train_scaled) // time_steps
remaining_data = len(X_train_scaled) % time_steps
num_samples_val = len(X_val_scaled) // time_steps
num_samples_test = len(X_test_scaled) // time_steps

# Reshape the input data for LSTM
X_train_lstm = X_train_scaled[:num_samples * time_steps].reshape(num_samples, time_steps, X_train_scaled.shape[1])
y_train_lstm = y_train_scaled[:num_samples * time_steps].reshape(num_samples, time_steps, y_train_scaled.shape[1])
X_val_lstm = X_val_scaled[:num_samples_val * time_steps].reshape(num_samples_val, time_steps, X_val_scaled.shape[1])
y_val_lstm = y_val_scaled[:num_samples_val * time_steps].reshape(num_samples_val, time_steps, y_val_scaled.shape[1])
X_test_lstm = X_test_scaled[:num_samples_test * time_steps].reshape(num_samples_test, time_steps, X_test_scaled.shape[1])
y_test_lstm = y_test_scaled[:num_samples_test * time_steps].reshape(num_samples_test, time_steps, y_test_scaled.shape[1])
# Define the shape of input data
input_shape = (time_steps, X_train_scaled.shape[1])

# Define the Encoder-Decoder model
encoder_decoder = tf.keras.models.Sequential([
    LSTM(50, activation='relu', input_shape=input_shape),
    RepeatVector(time_steps),
    LSTM(50, activation='relu', return_sequences=True),
    TimeDistributed(Dense(50, activation='relu')),
    Flatten(),
    Dense(25, activation='relu'),
    Dense(1)
])

# Compile the model
encoder_decoder.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

```python
history_encoder= encoder_decoder.fit(X_train_lstm,y_train_lstm,epochs=100,batch_size=32,validation_data=(X_val_lstm,y_val_lstm),verbose=1)
```

```python
import matplotlib.pyplot as plt

# Plot training loss
plt.plot(history_encoder.history['loss'], label='Training Loss')
# Plot validation loss
plt.plot(history_encoder.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training MAE
plt.plot(history_encoder.history['mae'], label='Training MAE')
# Plot validation MAE
plt.plot(history_encoder.history['val_mae'], label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.legend()
plt.show()
```
